```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv("cleaned_data_v1.csv")

# Display first few rows
df.head()
```

```python
# Check dataset info
df.info()

# Check for missing values
df.isnull().sum()
```

```python
df.fillna("Unknown", inplace=True)  # Replace missing values with 'Unknown'
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Primary_App", hue="Gender", palette="Set2")
plt.title("Dating App Preferences by Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
# Count users using multiple apps
df["Multiple_Apps"] = df["Secondary_Apps"].apply(lambda x: "Yes" if x != "None" else "No")

plt.figure(figsize=(6,4))
sns.countplot(x=df["Multiple_Apps"], hue=df["Multiple_Apps"], palette="coolwarm", legend=False)
plt.title("Users with Multiple Dating Apps")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=10, kde=True, color="purple")
plt.title("Age Distribution of Dating App Users")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=df["Multiple_Apps"], hue=df["Multiple_Apps"], palette="coolwarm", legend=False)
plt.xticks(rotation=45)
plt.title("User Distribution by Location")
plt.show()
```
