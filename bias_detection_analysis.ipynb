```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv("cleaned_data_v1.csv")

# Display first few rows
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Gender", hue="Primary_App", palette="coolwarm")
plt.title("Gender Bias in Dating App Preferences")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Age", hue="Gender", kde=True, bins=10, palette="coolwarm")
plt.title("Age and Gender Representation in Dataset")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Location", hue="Location", palette=["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#e377c2"], legend=False)
plt.xticks(rotation=45)
plt.title("Users by Location")
plt.xlabel("Location")
plt.ylabel("Number of Users")
plt.show()

```
